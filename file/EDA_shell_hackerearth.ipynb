```python
import pandas as pd
import numpy as np

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = "https://raw.githubusercontent.com/rohitgupta29/hack_shell/main/Data/power_curve.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
#Power Curve Anaylsis
#Theoretical power curve
plt.scatter(df['Wind Speed (m/s)'],df['Power (MW)'])
plt.ylabel('Power (MW)')
plt.xlabel('Wind speed (m/s)')
plt.grid(True)
plt.legend(['Power'], loc='upper left')
plt.show()
```

## Wind Speed higher than 25 m/s makes the power generation almost zero.

```python
#Power Curve Anaylsis
#Theoretical power curve
plt.scatter(df['Wind Speed (m/s)'],df['Thrust Coeffecient'])
plt.ylabel('Thrust Coeffecient')
plt.xlabel('Wind speed (m/s)')
plt.grid(True)
plt.legend(['Thrust coeffecient'], loc='upper left')
plt.show()
```

## Thrust coefficient works between wind speed of around 5 to 25 m/s
